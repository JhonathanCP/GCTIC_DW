```python
import oracledb
import pandas as pd
from sqlalchemy import create_engine
import sys
from sqlalchemy import create_engine
from sqlalchemy import text
import sys
from decouple import config

DB_USER=config("USER2_BDI_POSTGRES")
DB_PASSWORD=config("PASS2_BDI_POSTGRES")
DB_NAME="dw_essalud"
DB_PORT="5432"
DB_HOST=config("HOST2_BDI_POSTGRES")
cadena4  = f"postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
engine4 = create_engine(cadena4)
connection4 = engine4.connect()
```

```python
pdcamas = pd.read_sql_query(f"SELECT * FROM {camas}", con=connection4)
pdhabitaciones = pd.read_sql_query(f"SELECT * FROM {habitaciones}", con=connection4)
pdestreg = pd.read_sql_query(f"SELECT cod_est, des_est FROM {estreg}", con=connection4)
pdtiphoscam = pd.read_sql_query(f"SELECT cod_tipcam, des_tipcam FROM {tiphoscam}", con=connection4)
pdesthoscam = pd.read_sql_query(f"SELECT cod_estcam, des_estcam FROM {esthoscam}", con=connection4)
pdoricas = pd.read_sql_query(f"SELECT ori_cod, ori_nom FROM {oricas}", con=connection4)
pdred = pd.read_sql_query(f"SELECT id_red,cod_red,des_red FROM {red}", con=connection4)
pdcas = pd.read_sql_query(f"SELECT id_red,cod_cas,des_cas, ori_cas FROM {cas} where id_red is not null", con=connection4)
pdservicios = pd.read_sql_query(f"SELECT cod_ser, des_ser FROM {servicios}", con=connection4)
pdestenf = pd.read_sql_query(f"SELECT ori_cas, cod_cas as estenf_cod_cas, are_cod as estenf_are_cod, ser_cod as estenf_serv_cod, est_cod, est_des FROM {estenf}", con=connection4)
pdareas = pd.read_sql_query(f"SELECT cod_are,des_are FROM {areas}", con=connection4)
pdtiphoshab = pd.read_sql_query(f"SELECT cod_tiphab, des_tiphab FROM {tiphoshab}", con=connection4)
pdesthoshab = pd.read_sql_query(f"SELECT cod_esthab, des_esthab FROM {esthoshab}", con=connection4)
```
