```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pytesseract
from PIL import Image
import requests
from io import BytesIO
import cv2
import numpy as np
import requests
import time
```

```python
driver = webdriver.Chrome()
# URL de la página web
url = 'https://app1.susalud.gob.pe/registro/'

# Abrir la página en el navegador
driver.get(url)
```

```python
dni_input = driver.find_element(By.ID, 'txtNroDoc')
dni_input.send_keys('70189536')  # Reemplaza '12345678' con tu número de DNI

fec_input = driver.find_element(By.ID, 'txtFechaNac')
fec_input.send_keys('23/07/1997')  # Reemplaza '12345678' con tu número de DNI

captcha_img = driver.find_element(By.ID, 'divCaptcha').find_element(By.TAG_NAME, 'img')

# Tomar una captura del elemento de la imagen
captcha_image = captcha_img.screenshot_as_png

# Convertir la captura en un objeto de imagen utilizando PIL
captcha_image = Image.open(BytesIO(captcha_image))

text_captcha = pytesseract.image_to_string(captcha_image, lang='eng', config='--psm 10 --oem 3 -c tessedit_char_whitelist=0123456789ABCDEFGHIJKLMÑNOPQRSTUVWXYZ')

captcha = driver.find_element(By.ID, 'txtCaptcha')
captcha.send_keys(text_captcha)  # Reemplaza '12345678' con tu número de DNI

btn_consultar = driver.find_element(By.ID, 'btnConsultar')
btn_consultar.click()
```
