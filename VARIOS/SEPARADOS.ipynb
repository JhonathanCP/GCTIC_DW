```python
import pandas as pd
import re
```

```python
archivo_txt = '/home/ugadingenieria01/Documentos/GCTIC/VARIOS/USUARIOSSGD.csv'
usuarios = pd.read_csv(archivo_txt, delimiter=';', header=0, encoding='utf-8')
```

```python
def extraer_apellidos_nombres(texto):
    # Expresión regular para detectar apellidos compuestos como "DE LA CRUZ" o "DEL AGUILA"
    patron_compuesto = re.compile(r'\b(LA|DE LA|DEL|DE)\s+(\S+)\b')
    
    # Buscar todas las coincidencias de apellidos compuestos
    coincidencias = patron_compuesto.findall(texto)
    
    # Inicializar listas para apellidos y nombres
    apellidos = []
    nombres = []
    
    # Si hay apellidos compuestos, procesarlos
    for apellido_compuesto in coincidencias:
        apellidos.append(' '.join(apellido_compuesto))
        # Eliminar apellidos compuestos del texto original
        texto = texto.replace(' '.join(apellido_compuesto), '')
    
    # Tomar las dos primeras palabras como apellidos si aún no se han identificado dos
    palabras = texto.split()
    while len(apellidos) < 2 and palabras:
        apellidos.append(palabras.pop(0))
    
    # El resto del texto son los nombres
    nombres = ' '.join(palabras).strip()
    
    return ' '.join(apellidos), nombres

# Aplicar la función a la columna "APELLIDOS Y NOMBRES"
usuarios[['APELLIDOS', 'NOMBRES']] = usuarios['APELLIDOS Y NOMBRES'].apply(extraer_apellidos_nombres).apply(pd.Series)
```

```python
# Escribir el DataFrame resultante a un archivo CSV
usuarios.to_csv('/home/ugadingenieria01/Documentos/GCTIC/VARIOS/SGD-USUARIOS.csv', index=False, sep=';', encoding='utf-8')
```
