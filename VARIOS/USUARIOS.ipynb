```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Configuración de la conexión
server = '10.0.0.129'  # Reemplaza con el nombre o dirección IP de tu servidor SQL Server
database = 'DB_GENERAL'  # Reemplaza con el nombre de tu base de datos
username = 'sa'  # Reemplaza con tu nombre de usuario
password = 'Essalud23**'  # Reemplaza con tu contraseña
driver = 'ODBC Driver 17 for SQL Server'
# Cadena de conexión
connection_string = f'mssql+pyodbc://{username}:{password}@{server}/{database}?driver={driver}'
```

```python
try:
    # Establecer la conexión con SQLAlchemy
    engine = create_engine(connection_string)

    # Especificar el esquema
    schema = 'jcardenas'

    # Construir la consulta SQL
    query = f'SELECT * FROM {schema}.H_TRABAJADOR'

    # Leer los resultados en un DataFrame de Pandas
    df = pd.read_sql(query, engine)

except Exception as e:
    print(f'Error de conexión: {e}')

finally:
    # No es necesario cerrar la conexión con SQLAlchemy
    pass
```

```python
archivo_txt = '/home/ugadingenieria01/Documentos/GCTIC/GCTIC-VARIOS/SGD/USUARIOSCHRISTIAN.csv'
usuarios = pd.read_csv(archivo_txt, delimiter=';', header=0, encoding='utf-8')
```

```python
usuarios['existe'] = usuarios["CORREO ELECTRONICO "].apply(lambda x: x.replace('@essalud.gob.pe', '')).isin(df['EMAIL'])
```

```python
# Eliminar '@' de los correos electrónicos en ambos DataFrames
df['usuario_sin_arroba'] = df['EMAIL']
usuarios['usuario_sin_arroba'] = usuarios['CORREO ELECTRONICO '].apply(lambda x: x.replace('@essalud.gob.pe', ''))
```

```python
dfnuevo = pd.DataFrame()
dfnuevo['usuario_sin_arroba'] = df['usuario_sin_arroba']
dfnuevo['coddep_siad'] = df['coddep_siad'].astype(int)
dfnuevo['DNI_SISTEMA'] = df['DNI']
```

```python
usuarios = pd.merge(usuarios, dfnuevo, on="usuario_sin_arroba", how="left")
```

```python
usuarios.to_csv('usuariosSGD.csv', index=False)
```
