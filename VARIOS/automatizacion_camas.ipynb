```python
import imaplib
import email

# Configura los parámetros de conexión IMAP para Outlook 365
mail_settings = {
    "user": "gctic.gsit06@essalud.gob.pe",
    "password": "EssaludG23",
    "host": "wiracocha.essalud.com",
    "port": 993,  # Puerto IMAP seguro (IMAPS)
    "ssl": True
}

# Inicia la conexión IMAP
connection = imaplib.IMAP4_SSL(mail_settings["host"], mail_settings["port"])

# Inicia sesión en el servidor IMAP
connection.login(mail_settings["user"], mail_settings["password"])

# Selecciona la bandeja de entrada
connection.select('inbox')

# Busca los últimos 10 correos electrónicos
typ, data = connection.search(None, 'ALL')
for num in data[0].split()[-10:]:
    typ, msg_data = connection.fetch(num, '(RFC822)')
    raw_email = msg_data[0][1]
    email_message = email.message_from_bytes(raw_email)
    print('Asunto:', email_message['subject'])
    print('De:', email_message['from'])
    print('Fecha:', email_message['date'])

# Cierra la conexión IMAP
connection.logout()
```

```python
from onedrivesdk import AuthProvider, Client
from onedrivesdk.helpers import GetAuthCodeServer
from onedrivesdk.helpers.resource_discovery import ResourceDiscoveryRequest

# Configurar la autenticación
client_id = 'TU_CLIENT_ID'
client_secret = 'TU_CLIENT_SECRET'
redirect_uri = 'http://localhost:8080'

# Obtener el código de autorización
client_auth = AuthProvider(client_id, client_secret, redirect_uri)
auth_url = client_auth.get_auth_url(redirect_uri)
code = GetAuthCodeServer.get_auth_code(auth_url, redirect_uri)

# Iniciar sesión
client = Client(client_auth)
client.auth_provider.authenticate(code, redirect_uri, client_secret)

# Obtener el servicio de OneDrive
service_info = ResourceDiscoveryRequest().get_service_info(client)
service_root_url = service_info[0].service_resource_id

# Listar archivos en una carpeta específica
folder_id = 'ID_DE_LA_CARPETA'
items = client.item(drive='me', id=folder_id).children.get()

# Imprimir los nombres de los archivos
for item in items:
    print(item.name)
```
