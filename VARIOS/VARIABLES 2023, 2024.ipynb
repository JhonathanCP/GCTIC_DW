```python
df = pd.concat([df1,df2], ignore_index=True)
```

```python
df['ETAPA DE VIDA'] = np.nan
df.head()
```

```python
# Define conditions and replacement values
conditions = [
    (df['EDAD'] >= 0) & (df['EDAD'] <= 11),
    (df['EDAD'] >= 12) & (df['EDAD'] <= 17),
    (df['EDAD'] >= 18) & (df['EDAD'] <= 29),
    (df['EDAD'] >= 30) & (df['EDAD'] <= 59),
    (df['EDAD'] >= 60)
]
values = [
    'GE01: Niño',
    'GE02: Adolescente',
    'GE03: Joven',
    'GE04: Adulto',
    'GE05: Adulto Mayor'
]

# Replace NaN values in 'ETAPA DE VIDA' based on 'EDAD' conditions
df['ETAPA DE VIDA'] = np.where(pd.isnull(df['ETAPA DE VIDA']),
                               np.select(conditions, values, default='Unknown'),
                               df['ETAPA DE VIDA'])
```
