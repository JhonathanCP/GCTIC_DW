```python
import pandas as pd
from decouple import config
from sqlalchemy import create_engine
import pandas as pd

DB_USER=config("USER2_BDI_POSTGRES")
DB_PASSWORD=config("PASS2_BDI_POSTGRES")
DB_NAME="INDICADORES_ESSALUD_2"
DB_PORT="5432"
DB_HOST=config("HOST2_BDI_POSTGRES")
cadena2  = f"postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
engine2 = create_engine(cadena2)
connection2 = engine2.connect()

redes_df = pd.read_sql_query("SELECT id_red, cod_red, des_red FROM dim_red", con=connection2)
ipress_df = pd.read_sql_query("SELECT id_cas, cod_red, des_cas, niv_cas, cod_cas FROM dim_cas", con=connection2)
```

```python
ipress_df = pd.merge(ipress_df, redes_df, how='left', on='cod_red')
```

```python
import requests


# Datos de inicio de sesión
login_url = 'http://10.0.28.15:3000/auth/login'
login_data = {
    'username': 'admin',  # Cambia esto por tu nombre de usuario
    'password': '365298',  # Cambia esto por tu contraseña
    'systemcode': 'SAD'
}

# Realizar la solicitud de inicio de sesión
login_response = requests.post(login_url, json=login_data, verify=False)

if login_response.status_code == 200:
    token = login_response.json()['token']
    print('Login exitoso. Token obtenido.')
else:
    print('Error en el login:', login_response.json())
    exit()

headers = {
    'x-access-token': token,
    'Content-Type': 'application/json'
}

# URLs para las solicitudes de creación
red_url = 'http://10.0.28.15:8000/red/'
ipress_url = 'http://10.0.28.15:8000/ipress/'


# Crear MainDependencies
for index, row in redes_df.iterrows():
    main_dependency_data = {
        'name': row['des_red'],
        'cod_essi' : row['cod_red'],
    }
    response = requests.post(red_url, json=main_dependency_data, headers=headers, verify=False)
    if response.status_code == 201:
        print(f"REDES  creada exitosamente.")
    else:
        print(f"Error al crear REDES :", response.json())

# Crear Dependencies
for index, row in ipress_df.iterrows():
    dependency_data = {
        'name': row['des_cas'],
        'cod_essi' : row['cod_cas'],
        'level' : row['niv_cas'],
        'RedId': row['id_red']
    }
    response = requests.post(ipress_url, json=dependency_data, headers=headers, verify=False)
    if response.status_code == 201:
        print(f"IPRESS creada exitosamente.")
    else:
        print(f"Error al crear IPRESS :", response.json())
```
