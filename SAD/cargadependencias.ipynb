```python
import pandas as pd
# Leer archivos CSV
main_dependency_df = pd.read_csv('maindependency.csv', encoding='utf-8', sep=';')
dependency_df = pd.read_csv('dependency.csv', encoding='utf-8', sep=';')
```

```python
import requests


# Datos de inicio de sesión
login_url = 'http://10.0.28.15:3000/auth/login'
login_data = {
    'username': 'admin',  # Cambia esto por tu nombre de usuario
    'password': '365298',  # Cambia esto por tu contraseña
    'systemcode': 'SAD'
}

# Realizar la solicitud de inicio de sesión
login_response = requests.post(login_url, json=login_data, verify=False)

if login_response.status_code == 200:
    token = login_response.json()['token']
    print('Login exitoso. Token obtenido.')
else:
    print('Error en el login:', login_response.json())
    exit()

headers = {
    'x-access-token': token,
    'Content-Type': 'application/json'
}

# URLs para las solicitudes de creación
main_dependency_url = 'http://10.0.28.15:8000/maindependency/'
dependency_url = 'http://10.0.28.15:8000/dependency/'


# Crear MainDependencies
for index, row in main_dependency_df.iterrows():
    main_dependency_data = {
        'name': row['name']
    }
    response = requests.post(main_dependency_url, json=main_dependency_data, headers=headers, verify=False)
    if response.status_code == 201:
        print(f"MainDependency '{row['name']}' creada exitosamente.")
    else:
        print(f"Error al crear MainDependency '{row['name']}':", response.json())

# Crear Dependencies
for index, row in dependency_df.iterrows():
    dependency_data = {
        'name': row['name'],
        'MainDependencyId': row['MainDependencyId']
    }
    response = requests.post(dependency_url, json=dependency_data, headers=headers, verify=False)
    if response.status_code == 201:
        print(f"Dependency '{row['name']}' creada exitosamente.")
    else:
        print(f"Error al crear Dependency '{row['name']}':", response.json())
```
