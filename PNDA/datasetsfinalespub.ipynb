```python
# Asegurándonos de que las columnas de fechas sean cadenas
def ensure_string_format(df, columns):
    for column in columns:
        df[column] = df[column].astype(str)
    return df

# Definiendo los dataframes y las columnas de fechas
dataframes = {
    # 'diabetes': diabetes,
    # 'enfermedad_renal': enfermedad_renal,
    'hiperlipidemia': hiperlipidemia,
    # 'hipertension': hipertension,
    # 'obesidad': obesidad
}

fecha_columnas = ['FECHA_MUESTRA', 'FEC_RESULTADO_1', 'FEC_RESULTADO_2']
fecha_inicio = '20200101'
fecha_fin = '20240501'

# Aplicando el filtro a todos los dataframes
for nombre, df in dataframes.items():
    df = ensure_string_format(df, fecha_columnas)
    df = df[df['FECHA_MUESTRA'] >= fecha_inicio]
    df = df[df['FEC_RESULTADO_1'] >= fecha_inicio]
    df = df[df['FEC_RESULTADO_2'] >= fecha_inicio]
    df = df[df['FECHA_MUESTRA'] < fecha_fin]
    df = df[df['FEC_RESULTADO_1'] < fecha_fin]
    df = df[df['FEC_RESULTADO_2'] < fecha_fin]
    dataframes[nombre] = df

# Asignando los dataframes filtrados de vuelta a las variables originales
# diabetes = dataframes['diabetes']
# enfermedad_renal = dataframes['enfermedad_renal']
hiperlipidemia = dataframes['hiperlipidemia']
# hipertension = dataframes['hipertension']
# obesidad = dataframes['obesidad']
```
