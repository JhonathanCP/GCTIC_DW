```python
def process_id_paciente(value):
    if value.isdigit():  # Verificar si la cadena contiene solo dígitos
        value = str(int(value) * int(value)).encode()
    else:
        value = value.encode()
    compressed_value = zlib.compress(value)
    encoded_value = base64.b64encode(compressed_value).decode()
    return encoded_value
```

```python
pacientes['id_paciente'] = pacientes['doc_paciente'].apply(process_id_paciente)
```
