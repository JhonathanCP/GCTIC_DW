```python
import requests
import pandas as pd
```

```python
reportes = pd.read_csv('DATA-REPORTES.csv')
permisos = pd.read_csv('DATA-PERMISOS.csv')
usuarios = pd.read_csv('DATA-USUARIOS.csv')
```

```python
columnas_eliminar = ['group', 'module', 'name', 'description', 'version', 'active']

# Eliminar las columnas
reportes = reportes.drop(columns=columnas_eliminar)
```

```python
reportes = reportes[reportes['free']!=True]
```

```python
columnas_eliminar = ['free','limited','restricted']
reportes = reportes.drop(columns=columnas_eliminar)
```

```python
columnas_eliminar = ['id']
permisos = permisos.drop(columns=columnas_eliminar)
```

```python
permisosUsuarios = pd.merge(usuarios, permisos, on='username', how='inner')
```

```python
permisosUsuarios = pd.merge(permisosUsuarios, reportes, on='link', how='inner')
```

```python
columnas_eliminar = ['username', 'link']
permisosUsuarios = permisosUsuarios.drop(columns=columnas_eliminar)
permisosUsuarios = permisosUsuarios.rename(columns={'id_x': 'userId'})
permisosUsuarios = permisosUsuarios.rename(columns={'id_y': 'reportId'})
```

```python
import warnings
from urllib3.exceptions import InsecureRequestWarning

# Ignorar solo las advertencias de InsecureRequestWarning
warnings.simplefilter('ignore', InsecureRequestWarning)

import pandas as pd
import requests

# Datos de inicio de sesión
login_url = 'http://10.0.28.15:3000/auth/login'
login_data = {
    'username': 'admin',  # Cambia esto por tu nombre de usuario
    'password': '365298',  # Cambia esto por tu contraseña
    'systemcode': 'SAD'
}

# Realizar la solicitud de inicio de sesión
login_response = requests.post(login_url, json=login_data, verify=False)

if login_response.status_code == 200:
    token = login_response.json()['token']
    print('Login exitoso. Token obtenido.')
else:
    print('Error en el login:', login_response.json())
    exit()

# URL para crear usuarios
create_user_url = 'http://10.0.28.15:3000/user/add-report'

# Headers con el token de autenticación
headers = {
    'x-access-token': token,
    'Content-Type': 'application/json'
}

# Crear usuarios
for index, row in permisosUsuarios.iterrows():
    user_data = {
        'userId': int(row['userId']),  # Convertir a tipo int
        'reportId': int(row['reportId'])  # Convertir a tipo int
    }
    create_user_response = requests.post(create_user_url, headers=headers, json=user_data, verify=False)
    
    try:
        response_json = create_user_response.json()
    except ValueError:  # includes simplejson.decoder.JSONDecodeError
        response_json = create_user_response.text

    if create_user_response.status_code == 201:
        print(f"Usuario {row['userId']} y Modulo {row['reportId']} relacionado exitosamente.")
    else:
        print(f"Error al crear el permiso para {row['userId']} en Modulo {row['reportId']}: {response_json}")

print('Proceso completado.')
```
